Tokenisasi

```python
from nltk.tokenize import word_tokenize
import nltk
nltk.download('punkt')
nltk.download('punkt_tab')
```

```python
import pandas as pd
#Load data hasil scraping
df = pd.read_csv("/content/ulasan_aplikasi_bahasa_casefolded.csv")

#fungsi untuk melakukan tokenisasi
def tokenize_text(text):
    return word_tokenize(text)

#menambahkan kolom baru tokenize_ulasan
df['tokenize_review'] = df['casefolded_review'].apply(tokenize_text)

#menampilkan dataframe dengan kolom baru 'tokenize_ulasan
df.head()
```

Menghapus Tanda Baca

```python
import string

#fungsi untuk menghapus tanda baca
def remove_punctuation(tokens):
  table= str.maketrans("","",string.punctuation)
  return [word.translate(table) for word in tokens]

#menambahkan kolom baru tanpa tanda baca
df['review_tanpa_tanda_baca'] = df['tokenize_review'].apply(remove_punctuation)

#menampilkan dataframe dengan kolom baru
df.head()
```

Menghapus Whitespace (Karakter Kosong)

```python
#fungsi untuk membersihkan setiap token dari spasi dan tab di awal dan akhir
def clean_whitespace(tokens):
  return [token.strip() for token in tokens]

#menambahkan kolom baru tanpa spasi atau whitespace
df['review_tanpa_spasi'] = df['review_tanpa_tanda_baca'].apply(clean_whitespace)

df.head()
```

Stopwords (Menghapus Kata Yang Tidak Bermakna)

```python
import nltk
nltk.download('stopwords')

#mengambil semua kamus stopwords bahasa indonesia
stop_words = set(nltk.corpus.stopwords.words('indonesian'))

#fungsi untuk menghapus stopwords
def remove_stopwords(tokens):
  return [word for word in tokens if word not in stop_words]

#menambahkan kolom baru tanpa stopwords
df['review_tanpa_stopwords'] = df['normalized_review'].apply(remove_stopwords)

df.head()
```

Stemming

```python
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory

# Membuat objek stemmer
stemmer = StemmerFactory().create_stemmer()

# Fungsi untuk melakukan stemming
def stem_text(tokens):
    if not tokens: # Check if the list is empty
        return []
    stemmed_tokens = [stemmer.stem(token) for token in tokens]
    return stemmed_tokens

# Menerapkan stemming ke kolom review_tanpa_stopwords
df['stemmed_review'] = df['review_tanpa_stopwords'].apply(stem_text)

df.head()
```

Menghapus kata objek, angka, dan kurang dari 3 huruf

```python
# Definisikan daftar kata objek yang ingin dihapus
object_words_to_remove = ["duolingo", "hellotalk", "helo talk", "hello talk", "busuu", "busu", "yaa", "deh", "nya", "kek", "nih", "sih", "dah", "kan", "kak", "apa", "jir"] # Anda bisa menambahkan kata lain di sini

# Fungsi untuk menghapus kata-kata pendek, kata objek, dan angka
def remove_short_object_and_numbers(tokens):
    if not tokens:  # Check if the list is empty
        return []
    # Filter tokens based on length, not being in the object_words_to_remove list, and not being purely numeric
    cleaned_tokens = [token for token in tokens if len(token) >= 3 and token not in object_words_to_remove and not token.isdigit()]
    return cleaned_tokens

# Menerapkan fungsi ke kolom 'stemmed_review'
df['cleaned_review'] = df['stemmed_review'].apply(remove_short_object_and_numbers)

# Menampilkan dataframe dengan kolom baru
df.head()
```

**Visualisasi Bigram**

```python
from collections import Counter
import itertools
import matplotlib.pyplot as plt

# Menggabungkan semua token dari kolom 'cleaned_review' menjadi satu list
all_tokens = list(itertools.chain(*df['cleaned_review']))

# Membuat bigram dari list token
bigrams = list(nltk.bigrams(all_tokens))

# Menghitung frekuensi bigram
bigram_counts = Counter(bigrams)

# Mengambil 20 bigram yang paling umum
top_20_bigrams = bigram_counts.most_common(20)

# Memisahkan bigram dan hitungannya untuk visualisasi
bigram_words = [" ".join(bigram[0]) for bigram in top_20_bigrams]
bigram_counts = [bigram[1] for bigram in top_20_bigrams]

# Membuat bar plot untuk visualisasi bigram
plt.figure(figsize=(12, 8))
plt.barh(bigram_words, bigram_counts, color='skyblue')
plt.xlabel("Frekuensi")
plt.ylabel("Bigram")
plt.title("20 Bigram Paling Umum")
plt.gca().invert_yaxis() # Membalik urutan agar yang paling umum di atas
plt.show()
```

**Visualisasi Trigram**

```python
from collections import Counter
import itertools
import matplotlib.pyplot as plt

# Menggabungkan semua token dari kolom 'cleaned_review' menjadi satu list
all_tokens = list(itertools.chain(*df['cleaned_review']))

# Membuat trigram dari list token
trigrams = list(nltk.trigrams(all_tokens))

# Menghitung frekuensi trigram
trigram_counts = Counter(trigrams)

# Mengambil 20 trigram yang paling umum
top_20_trigrams = trigram_counts.most_common(20)

# Memisahkan trigram dan hitungannya untuk visualisasi
trigram_words = [" ".join(trigram[0]) for trigram in top_20_trigrams]
trigram_counts = [trigram[1] for trigram in top_20_trigrams]

# Membuat bar plot untuk visualisasi trigram
plt.figure(figsize=(12, 8))
plt.barh(trigram_words, trigram_counts, color='salmon')
plt.xlabel("Frekuensi")
plt.ylabel("Trigram")
plt.title("20 Trigram Paling Umum")
plt.gca().invert_yaxis() # Membalik urutan agar yang paling umum di atas
plt.show()
```
